```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Analytics/a1_Dataset_10Percent.xlsx')
```

```python
data['DemAffl']=data['DemAffl'].fillna(data['DemAffl'].mode()[0])
data['DemAge'] = data['DemAge'].fillna(data['DemAge'].mean())
data['DemClusterGroup'] = data['DemClusterGroup'].fillna(data['DemClusterGroup'].mode()[0])
data['DemGender'] = data['DemGender'].fillna(data['DemGender'].mode()[0])
data['DemReg'] = data['DemReg'].fillna(data['DemReg'].mode()[0])
data['DemTVReg'] = data['DemTVReg'].fillna(data['DemTVReg'].mode()[0])
data['LoyalTime'] = data['LoyalTime'].fillna(data['LoyalTime'].mean())
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data['DemClusterGroup'] = le.fit_transform(data['DemClusterGroup'].astype('str'))
integer_mapping = {l: i for i, l in enumerate(le.classes_)}
print(integer_mapping)

data['DemGender'] = le.fit_transform(data['DemGender'].astype('str'))
integer_mapping = {l: i for i, l in enumerate(le.classes_)}
print(integer_mapping)

data['DemReg'] = le.fit_transform(data['DemReg'].astype('str'))
integer_mapping = {l: i for i, l in enumerate(le.classes_)}
print(integer_mapping)

data['DemTVReg'] = le.fit_transform(data['DemTVReg'].astype('str'))
integer_mapping = {l: i for i, l in enumerate(le.classes_)}
print(integer_mapping)

data['LoyalClass'] = le.fit_transform(data['LoyalClass'].astype('str'))
integer_mapping = {l: i for i, l in enumerate(le.classes_)}
print(integer_mapping)
```

```python
data.head()

```

```python
data.columns
```

```python
X = data.drop(columns=['ID', 'TargetBuy'])
y = data['TargetBuy']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(y_test.shape)
print(X_test.shape)
```

```python
X_train.shape, y_train.shape
```

```python
model = LogisticRegression(max_iter=2000, solver='liblinear')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import joblib
joblib.dump(model, '/content/drive/MyDrive/Colab Notebooks/Analytics/3.2 Classifier_LoyalCustomers')
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(pd.DataFrame(confusion_matrix(y_test, y_pred) , index=["True 0", "True 1"], columns=["Pred 0", "Pred 1"]))
```

```python
predictions = model.predict_proba(X_test)
print(predictions)
```

```python
data = data.copy()

df_x_test = pd.DataFrame(X_test)
X_test['ID'] = data.loc[X_test.index, 'ID']

df_predictions_prob = pd.DataFrame(predictions, columns=['Prob_0', 'Prob_1'])
df_test_dataset = pd.DataFrame(y_test).reset_index(drop=True)
df_test_dataset.columns = ['Actual Outcome']

dfx = pd.concat([X_test.reset_index(drop=True),
                 df_test_dataset.reset_index(drop=True),
                 df_predictions_prob.reset_index(drop=True)], axis=1)

cols = ['ID'] + [col for col in dfx.columns if col != 'ID']
dfx = dfx[cols]

dfx.to_excel('/content/drive/MyDrive/Colab Notebooks/Analytics/ModelOutput_10Percent.xlsx', index=False)

dfx.head()
```
